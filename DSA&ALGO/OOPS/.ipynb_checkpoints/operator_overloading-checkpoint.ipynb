```python
class Vechlie:
    def __init__(self, make, model, fare):
        self.make = make
        self.model = model
        self.fare = fare
    def __str__(self):
        return "the make of the car is {} and the model is og {} with a fare of {} per min".format(self.make,self.model,self.fare)
    
    def __add__(self, other): # this helped us to add the fare of 2 other objects
        return self.fare+other.fare
    
    
    def __sub__(self, other): # this helped us to add the fare of 2 other objects
        return self.fare-other.fare
```

```python
myobj1 = Vechlie("huyndai", 2017, 150)
myobj2 = Vechlie("tata", 2015, 50)
```

```python
print(myobj1)
```

```python
print(myobj2)
```

```python
myobj1 + myobj2
```

```python
myobj1 - myobj2
```
